```python
def rosen(x,y):
    return (1-x)**2+100*(y-x**2)**2
```

```python
fig = plot.figure()
ax = fig.gca(projection='3d')
s = 0.5  # Try s=1, 0.25, 0.1, or 0.05
X = np.arange(-2, 2.+s, s)   #Could use linspace instead if dividing
Y = np.arange(-5, 10.+s, s)   #evenly instead of stepping...
X, Y = np.meshgrid(X, Y)
Z =  rosen(X,Y) 
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,
         linewidth=0, antialiased=False)  #Try coolwarm vs jet
plot.show()
```

```python
import cloudpickle
import pygmo 

Dv = [10,20,50,100,500]

#jet= plot.get_cmap('jet')
#colors = iter(jet(np.linspace(0,1,5)))

gene = 50*np.array(Dv)  #number of generations

d = [2,10,30,100]


for g in gene:
    for dim in d:
            #print(str(g)+' Generations')    
        print('*******************************************************')  
        for D in Dv:
            generation = g #D*50
            from pygmo import *     
            algo = algorithm(pso(gen = int(generation), 
                                 omega = 0.7298, # inertia weight (or constriction factor)
                                 eta1 = 1.5, #magnitude of the force, applied to the particles velocity, in the direction of its previous best position
                                 eta2 = 1.5, #magnitude of the force, applied to the particles velocity, in the direction of the best position in its neighborhood
                                 max_vel = 0.5, #maximum allowed particle velocity (as a fraction of the box bounds) - (0,1]
                                 variant = 5, # algorithm variant to use (one of 1 .. 6) 
                                 neighb_type = 2, #swarm topology to use (one of 1 .. 4) [gbest, lbest, Von Neumann, adaptive random]
                                 neighb_param = 4, #the neighbourhood parameter. If the lbest topology is selected (neighb_type=2), 
                                 #it represents each particles indegree (also outdegree) in the swarm topology. 
                                 #Particles have neighbours up to a radius of k = neighb_param / 2 in the ring. 
                                 #If the Randomly-varying neighbourhood topology is selected (neighb_type=4), 
                                 #it represents each particles maximum outdegree in the swarm topology. 
                                 #The minimum outdegree is 1 (the particle always connects back to itself). 
                                 #If neighb_type is 1 or 3 this parameter is ignored.
                                 memory = False, #when true the particle velocities are not reset between successive calls to evolve
                                 seed = random.randint(1,101)))

            algo.set_verbosity(5)
            prob = problem(rosenbrock(dim))
            pop = population(prob=prob, size=D, seed=None)  
            pop = algo.evolve(pop)
            best_fitness = pop.get_f()[pop.best_idx()]  
            uda = algo.extract(pso)
            log = uda.get_log()   
            plot.semilogy([entry[0] for entry in log],[entry[2]for entry in log]) #, 'k--',color=next(colors))  
            print (tabulate([['Pop Size(D):', str(D), 'Best fitness:', str(best_fitness)]]))

        plot.legend(Dv)
        plot.title('Dim = ' + str(dim))
        plot.ylabel('Objective Function')
        plot.xlabel('Number of Generations') 
        plot.show() 
        plot.savefig('rosen')
    
```
