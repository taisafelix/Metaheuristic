```python
def rosen(x,y):
    return (1-x)**2+100*(y-x**2)**2
```

```python
fig = plot.figure()
ax = fig.gca(projection='3d')
s = 0.5  # Try s=1, 0.25, 0.1, or 0.05
X = np.arange(-2, 2.+s, s)   #Could use linspace instead if dividing
Y = np.arange(-5, 10.+s, s)   #evenly instead of stepping...
X, Y = np.meshgrid(X, Y)
Z =  rosen(X,Y) 
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,
         linewidth=0, antialiased=False)  #Try coolwarm vs jet
plot.show()
```

## Study the effect of population size and function dimension: 

```python
from scipy.optimize import rosen
from tabulate import tabulate
```

```python
import cloudpickle
import pygmo 

Dv = [10,20,50,100,500]

#jet= plot.get_cmap('jet')
#colors = iter(jet(np.linspace(0,1,5)))

gene = 50*np.array(Dv) #number of generations
d = [2,10,30,100]
for g in gene:
    for dim in d:
            #print(str(g)+' Generations')    
            print('\n*******************************************************\n')  
            for D in Dv:
                generation = g #D*50
                from pygmo import *
                algo = algorithm(sga(gen = int(generation), 
                                     cr = .90, 
                                     eta_c = 1., 
                                     m = 0.02, 
                                     param_m = 1., param_s = 2, 
                                     crossover = "exponential", 
                                     mutation = "polynomial", #try different operators q3
                                     selection = "tournament",
                                     seed = random.randint(1,101)))
                #PARAMETERS
                #mutation.RANDOM
                #Random mutation (width is set by the width argument in PyGMO.algorithm.sga)
                #mutation.GAUSSIAN
                #Gaussian mutation (bell shape standard deviation is set by the width argument in
                #PyGMO.algorithm.sga multiplied by the box-bounds width)
                #selection.ROULETTE
                #Roulette selection mechanism
                #selection.BEST20
                #Best 20% individuals are inserted over and over again
                #crossover.BINOMIAL
                #Binomial crossover
                #crossover.EXPONENTIAL
                #Exponential crossover


                algo.set_verbosity(2)
                prob = problem(rosenbrock(dim))
                pop = population(prob=prob, size=D, seed=None)  
                pop = algo.evolve(pop)
                best_fitness = pop.get_f()[pop.best_idx()]  
                uda = algo.extract(sga)
                log = uda.get_log()   
                plot.semilogy([entry[0] for entry in log],[entry[2]for entry in log]) #, 'k--',color=next(colors))  
                print (tabulate([['Pop Size(D):', str(D), 'Best fitness:', str(best_fitness)]]))

            plot.legend(Dv)
            plot.title('Dim = ' + str(dim))
            plot.ylabel('Objective Function')
            plot.xlabel('Number of Generations') 
            plot.show() 

            #print('Dim: '+ str(dim))
            print('crossover: ' + 'exponential')
            print('mutation: ' + 'polynomial')
            print('selection: ' + 'tournament')
```
